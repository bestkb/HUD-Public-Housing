```python
def get_geoids(lat, lon):
    geoid_2010 = None
    geoid_2020 = None
    
    
    # Try to get the 2010 vintage GEOID
    try:
        result_2010 = cg.coordinates(x=lon, y=lat, vintage='2010')
        if result_2010 and int(result_2010["Census Tracts"][0]['GEOID']):
            tract_2010 = int(result_2010["Census Tracts"][0]['GEOID'])
        if result_2010 and int(result_2010["2020 Census Blocks"][0]['GEOID']):
            block_2010 = int(result_2010["2020 Census Blocks"][0]['GEOID'])
    except Exception as e:
        pass  # Handle any errors by skipping this entry
    
    # Try to get the 2020 vintage GEOID
    try:
        result_2020 = cg.coordinates(x=lon, y=lat, vintage='2020')
        if result_2020 and int(result_2020["Census Tracts"][0]['GEOID']):
            tract_2020 = int(result_2020["Census Tracts"][0]['GEOID'])
        if result_2020 and int(result_2020["2020 Census Blocks"][0]['GEOID']):
            block_2020 = int(result_2020["2020 Census Blocks"][0]['GEOID'])

    except Exception as e:
        pass  # Handle any errors by skipping this entry
    
    return pd.Series([tract_2010, tract_2020,block_2010,block_2020])
```
